```python
# Function show info, shape, nunique of dataframe
def info_shape_df(df):
    """Function show info, shape, nunique of dataframe.
    """
    print('Info DataFrame: ')
    df.info()
    print(f'Total Row DF:\n {df.shape}')
    print(f'Unique Value OF DataFrame:\n{df.nunique()}')
```

**What is the average interest rate on a large loan? What is the interest rate for a small loan?**  
I would consider large loans to be worth more than the average LoanOriginalAmount , and small loans to be worth less than the average LoanOriginalAmount.

```python
# Get mean LoanOriginalAmount
loan_mean = df.LoanOriginalAmount.mean()
loan_mean
```

```python
# Get large loan
loan_large = df[df['LoanOriginalAmount'] > loan_mean]
loan_large.shape
```

```python
# Average interest rate on a large loan
loan_large.BorrowerRate.mean()
```

```python
# Get small loan
loan_small = df[df['LoanOriginalAmount'] < loan_mean]
loan_small.shape
```

```python
# Average interest rate on a small loan
loan_small.BorrowerRate.mean()
```

```python
sum_rate = loan_large.BorrowerRate.mean() + loan_small.BorrowerRate.mean()
sum_rate
```

```python
deviant = loan_small.BorrowerRate.mean() - loan_large.BorrowerRate.mean()
(deviant/sum_rate)*100
```

**It is easy to see that small loans have higher interest rates than large loans.
The difference is about 0.12 or 12%.**
